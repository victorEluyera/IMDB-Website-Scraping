```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
```

```python
import requests
 
url = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'  # Replace this with the target website URL
 
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.5026.0 Safari/537.36 Edg/103.0.1254.0"

 
response = requests.get(url,headers={"User-Agent": user_agent})
 

```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
titles = soup.select('a.ipc-title-link-wrapper')
movie_info = soup.select('span.sc-be6f1408-8')
ratings = soup.select('span.ipc-rating-star')
titles
```

```python
movies=[]



for title, info, rating in zip(titles, movie_info, ratings):
        movie_title = title.find('h3').text.strip()
        movie_year_or_duration = info.text.strip()
        imdb_rating = rating.text.strip()
        vote_count_element = rating.find('span', class_='ipc-rating-star--voteCount')
        if movie_year_or_duration.isdigit() and len(movie_year_or_duration) == 4:
            year = movie_year_or_duration
        elif 'h' in movie_year_or_duration or 'm' in movie_year_or_duration:
            duration = movie_year_or_duration
        else:
            label = movie_year_or_duration
        if "Rate" not in imdb_rating:
            movies.append({
                        'Title': movie_title,
                        'Year': year,
                        'Duration': duration,
                        'IMDb Rating': imdb_rating,
                        'Label': label
                    })

df = pd.DataFrame(movies)

df.tail(40)
```
